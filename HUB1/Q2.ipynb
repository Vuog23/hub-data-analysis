```python
def describe(data):

    categorical_columns = data.select_dtypes(include = 'object')

    count, unique, top, freq = [], [], [], []

    for i in categorical_columns:
        count.append(categorical_columns[i].count())
        unique.append(len(categorical_columns[i].unique()))
        top_value = categorical_columns[i].value_counts().idxmax()
        top.append(top_value)
        freq.append(categorical_columns[i].value_counts().max())

    describe_object = pd.DataFrame([count, unique, top, freq], columns = categorical_columns.columns, index = ['count', 'unique', 'top', 'freq'])

    numeric_columns = data.select_dtypes(include = 'int64')

    count, mean, std, min, onequarter, onehalf, threequarter, max  = [], [], [], [], [], [], [], []

    for i in numeric_columns:
        count.append(numeric_columns[i].count())
        mean.append(numeric_columns[i].mean())
        std.append(numeric_columns[i].std())
        min.append(numeric_columns[i].min())
        onequarter.append(numeric_columns[i].quantile(0.25))
        onehalf.append(numeric_columns[i].quantile(0.5))
        threequarter.append(numeric_columns[i].quantile(0.75))
        max.append(numeric_columns[i].max())

    describe_numeric = pd.DataFrame([count, mean, std, min, onequarter, onehalf, threequarter, max], columns = numeric_columns.columns, index = ['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'])

    return describe_numeric, describe_object
```

```python
describe_numeric, describe_categorical = describe(df)
```
